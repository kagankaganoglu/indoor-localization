```python
import torch
import torchvision.models as models
import torchvision.transforms as transforms
from PIL import Image
import os

# Load the pre-trained VGG model
model = models.vgg16(pretrained=True)
model.eval()  # Set the model to evaluation mode

# Define transformations to apply to the input images
transform = transforms.Compose([
    transforms.Resize((224, 224)),  # Resize the input images to 224x224 (VGG's input size)
    transforms.ToTensor(),           # Convert the images to PyTorch tensors
    transforms.Normalize(            # Normalize the images using the mean and standard deviation of the ImageNet dataset
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]
    )
])

# Directory containing the images
folder_path = "drive/My Drive/ENS491 Bitirme Projesi/Tests/fire_color/"

# Get all files in the directory
files = os.listdir(folder_path)

# Filter out only the image files
image_files = [f"{folder_path}/{file}" for file in files if file.endswith(('.jpg', '.jpeg', '.png', '.bmp', '.gif'))]

# Iterate over the image files and perform inference
for image_file in image_files:
    # Load the image
    img = Image.open(image_file)

    # Apply transformations
    img = transform(img).unsqueeze(0)  # Add a batch dimension

    # Perform inference
    with torch.no_grad():
        output = model(img)

    # Get the predicted class index with the highest probability
    _, predicted_index = torch.max(output, 1)

    # Get the class label corresponding to the predicted index
    class_label = predicted_index.item()

    # Print the predicted class label
    print(f"Image: {image_file}, Predicted class: {class_label}")
```

```python
import torch
import torchvision.models as models
import torchvision.transforms as transforms
from torchvision.utils import draw_bounding_boxes
from PIL import Image
import os

# Load the pre-trained VGG model
model = models.vgg16(pretrained=True)
model.eval()  # Set the model to evaluation mode

# Define transformations to apply to the input images
transform = transforms.Compose([
    transforms.Resize((224, 224)),  # Resize the input images to 224x224 (VGG's input size)
    transforms.ToTensor(),           # Convert the images to PyTorch tensors
    transforms.Normalize(            # Normalize the images using the mean and standard deviation of the ImageNet dataset
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]
    )
])

# Directory containing the images
folder_path = "drive/My Drive/ENS491 Bitirme Projesi/Tests/fire_color/"

# Get all files in the directory
files = os.listdir(folder_path)

# Filter out only the image files
image_files = [f"{folder_path}/{file}" for file in files if file.endswith(('.jpg', '.jpeg', '.png', '.bmp', '.gif'))]

# Iterate over the image files and perform inference
for image_file in image_files:
    # Load the image
    img = Image.open(image_file)

    # Apply transformations
    img_tensor = transform(img).unsqueeze(0)  # Add a batch dimension

    # Perform inference
    with torch.no_grad():
        output = model(img_tensor)

    # Get the predicted class index with the highest probability
    _, predicted_index = torch.max(output, 1)

    # Visualize the image with predicted bounding boxes
    result_img = draw_bounding_boxes(img_tensor.squeeze(0), boxes=None, labels=predicted_index.unsqueeze(0), class_names=['predicted_class'], linewidth=3)

    # Convert the tensor image to PIL format for display
    result_img_pil = transforms.ToPILImage()(result_img)

    # Display the image
    result_img_pil.show()
```

```python
import torch
import torchvision.models as models
import torchvision.transforms as transforms
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from PIL import Image
import os

# Load the pre-trained VGG model
model = models.vgg16(pretrained=True)
model.eval()  # Set the model to evaluation mode

# Define transformations to apply to the input images
transform = transforms.Compose([
    transforms.Resize((224, 224)),  # Resize the input images to 224x224 (VGG's input size)
    transforms.ToTensor(),           # Convert the images to PyTorch tensors
    transforms.Normalize(            # Normalize the images using the mean and standard deviation of the ImageNet dataset
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]
    )
])

# Directory containing the images
folder_path = "drive/My Drive/ENS491 Bitirme Projesi/Tests/fire_color/"

# Get all files in the directory
files = os.listdir(folder_path)

# Filter out only the image files
image_files = [f"{folder_path}/{file}" for file in files if file.endswith(('.jpg', '.jpeg', '.png', '.bmp', '.gif'))]

# Iterate over the image files and perform inference
for image_file in image_files:
    # Load the image
    img = Image.open(image_file)

    # Apply transformations
    img_tensor = transform(img).unsqueeze(0)  # Add a batch dimension

    # Perform inference
    with torch.no_grad():
        output = model(img_tensor)

    # Get the predicted class index with the highest probability
    _, predicted_index = torch.max(output, 1)

    # Get the class label corresponding to the predicted index
    class_label = predicted_index.item()

    # Visualize the image with predicted bounding boxes
    fig, ax = plt.subplots(1)
    ax.imshow(img)

    # Add a bounding box for each detected object
    # Modify this part according to your model's output structure
    # For VGG, we don't have bounding box predictions, so let's just visualize the whole image
    # For object detection models like YOLO or SSD, this would involve iterating over detected bounding boxes and drawing them
    # For demonstration purposes, let's draw a rectangle covering the entire image
    width, height = img.size
    rect = patches.Rectangle((0, 0), width, height, linewidth=2, edgecolor='r', facecolor='none')
    ax.add_patch(rect)

    # Set title and display the image
    ax.set_title(f"Predicted class: {class_label}")
    plt.show()
```

```python
import zipfile
import io
import os
from PIL import Image
import random

# Directory containing the zip file
zip_file_path = "drive/My Drive/ENS491 Bitirme Projesi/Tests/rgbd-scene-time-lapses.zip"

# Open the zip file
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    # Get a list of all image files in the zip file
    image_files = [file for file in zip_ref.namelist() if file.lower().endswith(('.jpg', '.jpeg', '.png', '.bmp', '.gif'))]
    # Choose a random subset of image files (e.g., 10 files)
    selected_files = random.sample(image_files, k=min(10, len(image_files)))
    # Initialize a list to store image files
    images = []
    # Iterate over selected image files
    for file_name in selected_files:
        # Open the file in the zip archive
        with zip_ref.open(file_name) as file:
            # Read the image data
            image_data = io.BytesIO(file.read())
            # Open the image using PIL
            image = Image.open(image_data)
            # Append the image to the list
            images.append(image)

# Inference
results = model(images)

# Results
results.print()
results.show()  # Display the re



```
